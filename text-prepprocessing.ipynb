<h2>2- Remove HTML Tags

```python
# Let's import "re" library first.

import re
```

```python
#Create a function to cleand the HTML tags

#This function only accepts strings as input 
#def remove_html_tags(text):
#    pattern=re.compile('<.*?>')
#    return pattern.sub("",text)

#Let's write a new functipn to solve the problem of accepting a list as input

def remove_html_tags(texts):
  cleaned_texts = []
  for text in texts:
    pattern = re.compile('<.*?>') 
    cleaned_text = pattern.sub('', text)
    cleaned_texts.append(cleaned_text)

  return cleaned_texts
```

```python
data=remove_html_tags(data)
```

<h2>3- Remove URL

```python
#Create a function to cleand the URL

#This function only accepts strings as input 
#def remove_url(text):
#    pattern=re.compile(r'https?://\S+|www\.\S+')
#    return pattern.sub("",text)

#Let's write a new functipn to solve the problem of accepting a list as input

def remove_url(texts):
  cleaned_texts = []
  for text in texts:
    pattern = re.compile(r'https?://\S+|www\.\S+')
    cleaned_text = pattern.sub('', text)
    cleaned_texts.append(cleaned_text)

  return cleaned_texts
```

```python
data=remove_url(data)
```

<h2>4- Remove punctuation

```python
#Let's import "string" library
import string
```

```python
string.punctuation
```

```python
exclude=string.punctuation
```

```python
#Create a function to cleand remove punctuation

#This function only accepts strings as input 
#def remove_punc(text):
#    for char in exclude:
#        text=text.replace(char,"")
#    return text

#Let's write a new functipn to solve the problem of accepting a list as input

def remove_punc(texts):
  cleaned_texts = []
  
  if isinstance(texts, list):
    for text in texts:
      cleaned_text = text
      for char in exclude:
        cleaned_text = cleaned_text.replace(char,"")
      cleaned_texts.append(cleaned_text)

  else:
    cleaned_text = texts
    for char in exclude:
      cleaned_text = cleaned_text.replace(char,"")
    cleaned_texts = cleaned_text
  
  return cleaned_texts
```

```python
data = remove_punc(data)
```

<h2>10- Lemmatization 

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```

```python
stemming(data)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

def lammatization(text):
    words=text.split()

    lemmetizer=WordNetLemmatizer()

    lemetized_word=[lemmetizer.lemmatize(word) for word in words]
    
    return lemetized_word
```

```python
lammatization(data)
```
